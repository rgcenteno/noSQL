PyMongo é a libraría oficial de Python para conectarse a MongoDB. A súa estrutura baséase en tres obxectos principais:

1. **MongoClient** → Conexión ao servidor/cluster
2. **Database** → Representa unha base de datos
3. **Collection** → Representa unha colección dentro dunha BD

### 1.1 MongoClient
Representa a conexión co servidor ou clúster MongoDB.

**Métodos destacados**:
- `list_database_names()` → devolve a lista de bases de datos dispoñibles.
- `get_database(nome)` ou `cliente[nome]` → obtén unha base de datos.
- `close()` → pecha a conexión.

### 1.2. Database
Obxecto que representa unha base de datos concreta.

**Métodos destacados**:
- `list_collection_names()` → lista todas as coleccións.
- `get_collection(nome)` ou `bd[nome]` → obtén unha colección.
- `create_collection(nome)` → crea unha colección.
- `drop_collection(nome)` → elimina unha colección.

### 1.3. Collection
Obxecto que representa unha colección.

**Métodos destacados**: 
- Crear
    - `insert_one(documento)`
    - `insert_many([docs])`

- Ler
    - `find(filtro, proxección)` → devolve un cursor (iterable)
    - `find_one(filtro)`

- Actualizar
    - `update_one(filtro, actualización)`
    - `update_many(filtro, actualización)`

- Eliminar
    - `delete_one(filtro)`
    - `delete_many(filtro)`

- Outros métodos útiles
    - `count_documents(filtro)` → conta documentos que cumpren un filtro.
    - `aggregate(pipeline)` → permite usar pipelines de agregación.
    - `distinct(campo)` → valores únicos dun campo.

### 1.4. Cursor
Obxecto retornado por `find()` que permite iterar sobre os resultados.

**Métodos destacados**:
- `limit(n)` → limita o resultado a `n` documentos.
- `sort(campo,orde)` → Ordena documentos segundo un `campo` no sentido indicado por `orde`.
- `skip(n)` → permite saltarse os `n` primeiros documentos.

## 3. Inserir documentos

### 3.1. Inserir un único documento
Para inserir un documento primeiro creamos un *documento*, que será un `dict` cos campos e valores a inserir. Para facelo empregaremos o método `insert_one()` da colección. Antes diso deberemos obter a *colección* a partir da *bd* e a *bd* a partir do cliente.

````python
# Obtemos a BD e a colección
bd = cliente["proba_pymongo"]
coleccion = bd["xente"]
coleccion

# Creamos o documento a inserir
documento = {
    "nome": "Adrián Blanco",
    "idade": 41,
    "profesion": "profesor"
}

# Chamamos á función `insertOne()` para realizar a inserción.
resultado = coleccion.insert_one(documento)

# Comprobamos se a operación se realizou con éxito
print(resultado.inserted_id)
````

### 3.2. Inserir varios documentos
O proceso para inserir varios documentos é moi semellante ao que vimos para un só documento. En vez de inserir un `dict` inserimos unha `lista` de `dict`.

````python
# Obtemos a BD e a colección
bd = cliente["proba_pymongo"]
coleccion = bd["xente"]
coleccion

# Creamos a lista de documentos a inserir
docs = [
    {"nome": "Ana", "idade": 20},
    {"nome": "Xosé", "idade": 25}
]

# Chamamos á función `insertOne()` para realizar a inserción.
resultado = coleccion.insert_many(docs)

# Comprobamos se a operación se realizou con éxito
print(resultado.inserted_ids)
````

## 4. Consultar documentos
As consultas realizanse de forma moi parecida á empregada no shell de mongo, cos métodos `find()` e `find_one()`.

### 4.1. Obter un documento
O método `find_one()` devolve un obxecto `dict`.

````python
# Obtemos a BD e a colección
bd = cliente["proba_pymongo"]
coleccion = bd["xente"]

# Obtemos o documento
doc = coleccion.find_one()

# Amosámolo
print(doc)
````

### 4.2. Obter varios documentos
De forma análoga, o método `find()` debolve un obxecto `cursor` que permite iterar polos distintos documentos recuperados.

````python
# Obtemos a BD e a colección
bd = cliente["proba_pymongo"]
coleccion = bd["xente"]

# Obtemos os documentos
docs = coleccion.find()

# Recorremos e amosamos os documentos
for doc in docs:
    print(doc)
````

### 4.3. Consulta con filtro
Os parámetros pasados a ambos métodos (`find()` e `find_one()`) constrúense exactamente igual que ao empregar o shell de mongo. Isto permite:
- Empregar filtros nas consultas

````python
# Obtemos a BD e a colección
bd = cliente["proba_pymongo"]
coleccion = bd["xente"]

# Obtemos os documentos filtrados 
docs = coleccion.find({"idade": {"$gt": 30}})

for doc in docs:
    print(doc)
````

### 4.4. Consulta con proxección
- Realizar proxeccións para só amosar determinados campos:

````python
# Obtemos a BD e a colección
bd = cliente["proba_pymongo"]
coleccion = bd["xente"]

# Obtemos os documentos 
docs = coleccion.find({}, {"_id": 0, "nome": 1}) 
for doc in docs:
    print(doc)
````

## 5. Actualización de documentos
As actualizacións realízanse de forma moi semellante ás insercións.
### 5.1. Actualizar un documento
Para actualizar un documento empregamos o método `update_one()`

````python
# Obtemos a BD e a colección
bd = cliente["proba_pymongo"]
coleccion = bd["xente"]

# Actualizamos o documento. Seleccionamos empregando un filtro e modificamos co operador `$set`
resultado = coleccion.update_one(
    {"nome": "Ana"},
    {"$set": {"idade": 22}}
)

# Amosamos o número de documentos modificados
print(f"Documentos modificados: {resultado.modified_count}")
````

Para actualizar varios documentos emprégase o método `update()`:

````python
# Obtemos a BD e a colección
bd = cliente["proba_pymongo"]
coleccion = bd["xente"]

# Actualizamos os documentos
resultado = coleccion.update_many(
    {"idade": {"$lt": 30}},
    {"$inc": {"idade": 1}}
)

# Amosamos o número de documentos modificados
print(f"Documentos modificados: {resultado.modified_count}")
````

## 6. Eliminar documentos
Para elminar documentos empréganse os métodos `update_one()` e `update_many()`.

### 6.1. Eliminar un documento
Emprégase o método `update_one()`. Como parámetro pásaselle un *json* cos filtros a aplicar para seleccionar o documento a eliminar. Borrarse a primeira coincidencia.

````python
# Obtemos a BD e a colección
bd = cliente["proba_pymongo"]
coleccion = bd["xente"]

# Eliminamos o documento
resultado = coleccion.delete_one({"nome": "Xosé"})

# comprobamos o resultado
print(f"Documentos eliminados: {resultado.deleted_count}")
````

### 6.2. Eliminar varios documentos
Para elminar varios documentos cunha soa instrución emprégase a función `update_many()`. Acepta os mesmos parámetros e diferénciase en que elimina todas as coincidencias, non só a primeira.

````python
# Obtemos a BD e a colección
bd = cliente["proba_pymongo"]
coleccion = bd["xente"]

# Eliminamos os documentos
resultado = coleccion.delete_many({"idade": {"$gt": 50}})

# comprobamos o resultado
print(f"Documentos eliminados: {resultado.deleted_count}")
````

## 7. Uso de Aggregation Pipeline en Python

O método é exactamente igual ao shell, pero pasando unha lista de dicionarios:

````python
# Obtemos a BD e a colección
bd = cliente["sample_mflix"]
coleccion = bd["movies"]

# Especificamos o pipeline de agregación
pipeline = [
    {"$match": {"genres": "Drama"}},
    {"$group": {"_id": "$year", "total": {"$sum": 1}}},
    {"$sort": {"total": -1}}
]

# Executamos a o método `aggregate` para realizar a consulta agregada
docs = bd["movies"].aggregate(pipeline)

# Percorremos o cursor amosando os resultados
for doc in docs:
    print(doc)
````

## 8. Crear índices
### Que é un índice?
Un **índice** é unha estrutura de datos adicional que permite **localizar documentos moito máis rápido** sen ter que percorrer toda a colección. 

Botémoslle un ollo a como a fectan os índices ao comportamento de **MongoDB**:
- Sen índice → MongoDB fai un **collection scan** (le todos os documentos).
- Con índice → MongoDB consulta o índice a vai **directamente aos documentos relevantes**.

### Cando ten sentido crear un índice?
É recomendable crear un índice sobre campos que:
- se usan frecuentemente en `find()`, `delete()`, `update()`.
- aparecen en filtros (`where`)
- se usan para ordenar (`sort`)
- se usan en combinación (índices compostos)

### Creación básica de índices
Para crear un índice con *pymongo* emprégase o método `create_index()`. Acepta os seguintes parámetros:
- Unha `string` co nome do campo (crearíase un índice ascendente).
- Un array que contén:
  - Unha `string` co nome do campo.
  - Un `int` para indicar se é ascendente(`1`) ou descendente (`-1`).
- Un array de tuplas (`campo`, `orde`) permitiríanos crear un **índice composto**. É útil se se realizan frecuentemente consultas sobre eses dous campos. A orde dos compos no índice composto é importante.

````python
# Obtemos a BD e a colección
bd = cliente["proba_pymongo"]
coleccion = bd["xente"]

# Creación dun índice ascendente no campo `nome`
coleccion.create_index("nome")

# Creación dun índice descendente no campo `idade`
coleccion.create_index([("idade", -1)])

# Creación dun índice composto
# coleccion.create_index([("cidade", 1), ("idade", 1)])

````

### Uso de índices
Unha vez creados os índices, para usalos basta simplemente con realizar consultas normais sobre os mesmos. Por exemplo, ao ter creado un índice sobre o campo `idade` a seguinte consultasería moito máis rápida, ao non ter que revisar todos os documentos da colección:

````python
# Obtemos a BD e a colección
bd = cliente["proba_pymongo"]
coleccion = bd["xente"]

# Facemos a consulta
resultado = coleccion.find({"idade": 23})

# Amosamos os resultados
for doc in resultado:
    print(doc)
````

En consultas con ordenación o índice e aínda máis importante. 
- Se non hai índice MongoDB ten que ordenar toda a colección en memoria. 
- Se o hai xa tería os datos ordenados diretamente.


````python
# Obtemos a BD e a colección
bd = cliente["proba_pymongo"]
coleccion = bd["xente"]

# Realizamos a consulta
resultado = coleccion.find().sort("idade", -1)

# Amosamos os resultados
for doc in resultado:
    print(doc)
````

### Outras operacións con índices
O método `index_information` permite ver os índices dunha colección.

````python
# Obtemos a BD e a colección
bd = cliente["proba_pymongo"]
coleccion = bd["xente"]

print(coleccion.index_information())
````

## 9. Tratamento de datos extraídos de MongoDB
Unha vez obtidos os datos desde MongoDB usando *PyMongo*, é importante decidir **como traballalos en Python*. MongoDB é unha base de datos orientada a documentos, polo que o tratamento posterior adoita facerse con estruturas dinámicas (dict, listas) ou ferramentas de análise como pandas.

### Creación de clases para almacenar documentos
É posible modelar os documentos de MongoDB mediante clases de Python (por exemplo con `dataclasses` ou librarías como **Pydantic**), o que achega:

- tipado máis claro  
- validación de datos  
- mellor mantemento en proxectos grandes  

Sen embargo, no noso contexto **non nos interesa empregar clases**, polos seguintes motivos:

- MongoDB non impón un esquema fixo, polo que modelar clases pode introducir rigidez innecesaria.
- Nos exemplos e prácticas céntramonos en:
  - consultas
  - análise exploratoria
  - agregacións
  - exportación de datos

O uso de `dict` é:
- máis flexible
- máis próximo ao modelo real de MongoDB
- máis axeitado para traballos de análise e Big Data

> Polo tanto, traballaremos directamente cos documentos como **dicionarios de Python**, que é o enfoque máis habitual en scripts, ETL e análise de datos.

### Visualizaciónde datos
Ao imprimir directamente un documento de MongoDB poden aparecer tipos especiais como `ObjectId`, `datetime` ou estruturas aniñadas, o que fai que a saída non sexa moi lexible.

**Visualización con `pprint`**

O módulo estándar pprint permite amosar os documentos de forma máis clara:

### 9.3. Conversión de coleccións ou consultas a pandas

Pandas é unha ferramenta excelente para **análise de datos**, estatísticas e visualización, pero **non é a mellor opción para transformacións complexas nin grandes volumes**.

O fluxo habitual é:

1. Executar unha consulta ou agregación en MongoDB  
2. Converter o resultado a unha lista de dicionarios  
3. Crear un `DataFrame` de pandas  

#### Proceso básico

````python
import pandas as pd

# Obtemos a BD e a colección
bd = cliente["proba_pymongo"]
coleccion = bd["xente"]

# Almacenamos a consulta nun `cursor`
cursor = coleccion.find()

# Pasamos o `cursor` ao tipo `list`
docs = list(cursor)

# Creamos o dataframe a partir dunha lista de dicts.
df = pd.DataFrame(docs)

# Se o comapo _id non se vai usar como identificador ´habitual convertilo a texto
df["_id"] = df["_id"].astype(str)

df
````

#### Nota importante sobre o uso de pandas

✔ **pandas é axeitado para**:
- análise exploratoria
- estatísticas
- visualización

✘ **non é axeitado para**:
- grandes volumes de datos
- transformacións complexas
- procesado distribuído

👉 As **transformacións deben facerse previamente**:
- en MongoDB usando `aggregate()`
- ou con Apache Spark (que se verá na sección correspondente)

---

### Exemplo: agregación da colección `zips` (sample_training)

Na base de datos `sample_training`, a colección `zips` contén información xeográfica e demográfica por código postal.

#### Obxectivo do exemplo

Obter:
- poboación total por estado (`state`)
- número de códigos postais por estado


````python
# Obtemos a BD e a colección
bd = cliente["sample_training"]
coleccion = bd["zips"]

# Construimos o pipeline
pipeline = [
    {
        "$group": {
            "_id": "$state",
            "poboacion_total": {"$sum": "$pop"},
            "num_zips": {"$sum": 1}
        }
    },
    {
        "$sort": {"poboacion_total": -1}
    }
]

# Realizamos a consulta agregada
cursor = coleccion.aggregate(pipeline)
````

Así:
- A agregación faise no servidor de MongoDB
- Pandas só recivve o resultado final agregado
O seguinte paso é convertir o resultado a pandas

````python
import pandas as pd

# Convertimos o cursor ao tipo list
docs = list(cursor)

# Creamos o dataframe
df = pd.DataFrame(docs)

# Renomeamos o campo _id a estado (o campo estado cambiouse a _id ao agrupar)
df.rename(columns={"_id": "estado"}, inplace=True)
df
````

Esta *DataFrame* podería ser usado, posteriormente, para:
- gráficas
- estadísticas
- exportación a outros formatos
- análise

````python
# Obtención das estadísticas
df.describe()
````

````python
import matplotlib.pyplot as plt

# ordenar para que a gráfica sexa máis clara
df_ordenado = df.sort_values("poboacion_total", ascending=False).head(10)

plt.figure(figsize=(10, 6))
plt.bar(df_ordenado["estado"], df_ordenado["poboacion_total"])
plt.xticks(rotation=90)

plt.title("Poboación total dos 10 estados máis poboados")
plt.xlabel("Estado")
plt.ylabel("Poboación total")

plt.tight_layout()
plt.show()
````
